### Loading csv file as dataframe


```python
csvDataFrame = pd.read_csv('cereals.csv')
csvDataFrame
```

```python
csvDataFrame.set_index('name')
```

```python
'''
BUT THIS WAS NOT SET INPLACE
'''
csvDataFrame
```

#### inplace

```python
'''
PERMANENT CHANGE
'''
csvDataFrame.set_index('name',inplace=True)
csvDataFrame
```

#### head() /tail()

```python
csvDataFrame.head(5)
```

```python
csvDataFrame.tail(5)
```

#### describe()

```python
csvDataFrame.describe()
```

#### row slicing

```python
csvDataFrame[5:8]
```

#### column indexing

```python
csvDataFrame[['calories','rating']]
```

#### boolean list

```python
csvDataFrame = csvDataFrame[0:5]
secondrow = [False,True,False,False,False]
csvDataFrame[secondrow]
```

#### filtering rows

```python
condition = csvDataFrame['calories']>70
print(condition)
csvDataFrame[condition]
```

```python
csvDataFrame[(csvDataFrame['calories']>70) & (csvDataFrame['rating']<40)]
```

```python
csvDataFrame[(csvDataFrame['calories']>70) | (csvDataFrame['rating']<40)]
```

#### loc()

```python
csvDataFrame.loc[0,'name']
```

```python
'''
ROW AND COLUMN AS LIST
'''
csvDataFrame.loc[[0],['name']]
```

```python
csvDataFrame.loc[0:5,'name': 'protein']
```

```python
csvDataFrame.loc[[0,5],'name': 'protein']
```

#### iloc

```python
'''
takes rows and column positions and not there labels
'''
print(csvDataFrame)
csvDataFrame.iloc[5,3]
```

```python
csvDataFrame.iloc[0:5,0:3]
```

```python
csvDataFrame.iloc[[2,4,6],0:2]
```

#### adding/deleting data

```python
csvDataFrame.loc[11] = ['Trix',110,1,25,27.435435]
csvDataFrame
```

```python
csvDataFrame.drop(11, axis=0, inplace=True)
csvDataFrame
```

```python
csvDataFrame['Extra'] = [str(x) for x in range(65,75)]
csvDataFrame
```

```python
csvDataFrame.drop('Extra', axis = 1, inplace=True)
csvDataFrame
```

```python
csvDataFrame.sort_values(by='calories')
```

```python
csvDataFrame.sort_values(by='calories', ascending=False)
```

#### exporting and saving csv

```python
csvDataFrame.to_csv('exported_csv', index_label=False)
new = pd.read_csv('exported_csv')
new
```

#### concat() : to concatenate two data frames

```python
pd.concat([csvDataFrame[0:3], csvDataFrame[5:8]], axis=0)
```

#### groupby()

```python
d1 = {'Gender': ['female','male','female','male'], 'Score':[85,88,95,80]}
df1 = pd.DataFrame(d1)
df1
```

```python
df1.groupby(df1['Gender']).mean()
```
