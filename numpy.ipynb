##### adding/subtracting/multiplying/dividing/to the power a scalar

```python
'''
Adding a Scalar value 2 in numpy array doesn't gives an error. 
But it would give an error, if it was a normal list 
'''
```

```python
d2array = np.array([[1,2,3],[4,5,6]])
print(d2array+2-2*3/3)
```

```python
print(d2array**2)
```

##### transpose

```python
print(d2array.T)
```

##### element wise addition/subtraction/multiplication/division 

```python
'''
IF normal list
+ -> concatenates the lists
-, /, *, ** -> will give error
'''
```

```python
matrix1 = np.array([[1,2,3],[4,5,6],[0,0,0]])
matrix2 = np.array([[7,8,9],[10,11,12],[0,0,0]])

print(matrix1 + matrix2, '\n')
print(matrix1 - matrix2, '\n')
print(matrix1 / matrix2, '\n')
print(matrix1 * matrix2, '\n')
print(matrix1 ** matrix2, '\n')
```

##### matrix multiplication

```python
np.matmul(matrix1,matrix2)
```

#### statistics

```python
print(matrix1)
print(np.min(matrix1))
print(np.max(matrix1))
print(np.sum(matrix1))
print(round(np.mean(matrix1),2)) 
print(round(np.std(matrix1),2)) # standard deviation rounded off to 2
print(np.median(matrix1))
```
