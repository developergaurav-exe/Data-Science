#### plotting on a same area

```python
x2_axis = [6,7,8,9]
y2_axis = [12,23,34,54]

x3_axis = [2,3,4,5]
y3_axis = [10,11,12,13]

plt.title('Multiple Graphs')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.plot(x1_axis, y1_axis,'r')
plt.plot(x2_axis, y2_axis, 'g')
plt.plot(x3_axis, y3_axis, 'b')

# To accomodate for space rotate x labels
plt.xticks(rotation = 90)

#ADDING LEGEND
plt.legend(['Regular','Linear','Exponential'])
```

### Plotting Histograms

```python
values = [10,15,20,10,15]

# Add color and width
plt.hist(values, color='y', width = 0.5)
```

### Plotting Bar Charts


```python
values = [5,10,15]
plt.bar(values, [0.5,1,2], width=2)
```

### Plotting Pie Chart


```python
plt.pie([5,10,20,7,16], labels=['a','b','c','d','e'], explode=[0,0,0.2,0,0.2], colors=['#B2BEB5','#7393B3','#6082B6','#899499','#E5E4E2'])
```

### Plotting Scatter Plots

```python
x = x1_axis + x2_axis
y = y1_axis + y2_axis

plt.title('Scatter Plot')
plt.xlabel('X-Axix')
plt.ylabel('Y-Label')

plt.scatter(x,y,color = 'r')
```

```python
'''
COLOR MAPS AND COLORS AVAILABLE
'''
plt.scatter(x,y,c = [1,2,3,4,5,6,7,8], cmap = 'Accent')
```

### Plotting polar Plots

```python
import numpy as np
theta = np.arange(0, (2*np.pi), 0.01) # generates a sequence of float
r = 10

for i in theta:
    plt.polar(i, r)
```

### Creating multiple subplots

```python
plt.subplot(1,2,1)
plt.plot(x1_axis, y1_axis, 'r')
plt.title('Loss')
plt.xlabel('x-label')
plt.ylabel('y-label')

plt.subplot(1,2,2)
plt.plot(x2_axis, y2_axis, 'g')
plt.title('Profit')
plt.xlabel('x-label')
plt.ylabel('y-label')

plt.tight_layout() # for space
```
