### Time Series
#### Squence of data measured over a regular time intervals
##### Analysing time series to identify a pattern in the data and develop a forecasting model

##### libraries for time series analysis:
- Pandas
- Numpy
- Matplotlib
- datetime

#### getting stocks data

```python
import yfinance as yf
df = yf.download('MSFT', start = '2022-06-01', end='2022-06-30')
df.head(10)
```

#### converting dataset to time series

```python
'''
Though its already in datetime format but see next cell for format
'''
type(df.index[0])
```

```python
import pandas as pd
df.index = pd.to_datetime(df.index, format = '%Y-%m-%d %I-%p')
print(type(df.index[0]))
df
```

###### As stock markets are closed on Saturdays and sundays, there will be no data available for them

#### working with time series

```python
jun18 = pd.to_datetime('2022, 6, 18')
jun18.day_name()
```

```python
'''
Max/Min date
'''

print(df.index.max())
print(df.index.min())
```

```python
'''
Datetime range
'''
df.index.max() - df.index.min()
```

```python
'''
Time based indexing
'''

df.loc['2022-06-29']
```

```python
df.loc['2022-06']
```

```python
df.loc['2022-06-1':'2022-06-14']
```

#### Visualizing a time series

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.title('Microsoft Stocks in June 2022')
plt.xlabel('Business Dates')
plt.ylabel('Price')

plt.plot(df['Adj Close'])
```

```python
df1 = yf.download('MSFT', start = '2021-01-01', end='2021-12-31')
df1
```

```python
plt.figure(figsize=(15,5))

plt.title('Microsoft Stock in year 2021')
plt.ylabel('Price')
plt.xlabel('Months')

plt.plot(df1['Adj Close'])
```

```python
plt.figure(figsize=(15,5))

plt.title('Microsoft Stock in Dec 2021')
plt.ylabel('Price')
plt.xlabel('Business Days')

plt.plot(df1.loc['2021-12']['Adj Close'])
```
